# Data Visualization using matplotlib

## Basic Program to plot a graph with the given coordinates

```python
import matplotlib.pyplot as plt
import numpy as np

# x and y pts for coordinates

x = [0,10]
y = [0,125]

# Plotting using the pyplot submodule

plt.plot(x,y)
plt.show()
```

## Plotting a line with DataFrame

```python
import matplotlib.pyplot as plt
import pandas as pd

# Adding the data
data = {
    "Cricket Bat":["SG","BDM","SS","GM","Kookabura","Spartan"],
    "MRP":[2000,2200,2400,2600,2800,3000],
    "Weight_Grams":[1100,1200,1250,1330,1480,1600]
}

# Adding the above data to the dataframe
dataFrame = pd.DataFrame(data)

# Plotting a line using the dataframe where its columns will be our coordinates
plt.plot(dataFrame["MRP"],dataFrame["Weight_Grams"])
plt.show()
```

## Adding Grid Lines, Labels and Plot titles to a plot

```python
import matplotlib.pyplot as plt
import pandas as pd

# Adding the data
data = {
    "Cricket Bat":["SG","BDM","SS","GM","Kookabura","Spartan"],
    "MRP":[2000,2200,2400,2600,2800,3000],
    "Weight_Grams":[1100,1200,1250,1330,1480,1600]
}

# Adding the above data to the dataframe
dataFrame = pd.DataFrame(data)

# Plotting a line using the dataframe where its columns will be our coordinates
plt.plot(dataFrame["MRP"],dataFrame["Weight_Grams"])

plt.grid()
plt.xlabel("MRP")
plt.ylabel("Weight_Grams")
plt.title("Cricket Bat")  #By default, it is center. To place it in a diff location use loc = 'right/left'

plt.show()
```

## Adding a legend to the graph. (Functionalities-position, background colour, font size)

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(5)  #Using arrange function for an evenly spaced array
b = [2,4,6,8,10]
c = [5,6,7,8,9]

fig = plt.figure()  #for new figure
ax = plt.subplot()  # for the axis of a given figure

ax.plot(a,b, 'k--', label = 'Frequency')
ax.plot(a,c, 'k:', label = 'Periods')

legend = ax.legend(loc = "upper center", fontsize = 'xx-large') #Available positions are upper left, upper right, lower left, lower right, upper center, lower center, center left, center right and center

legend.get_frame().set_facecolor('red')
legend.get_frame().set_edgecolor('green')

plt.title("Frequency of a Signal")
plt.xlabel("Time")
plt.ylabel("Frequency")

plt.show()
```

## Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np

student = np.array(["Tim","John","Jacob","Karl","Gary","Mike"])
marks = np.array([85,59,99,77,67,95])

plt.bar(student,marks)

plt.title("Marks of Students")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

## Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np

cricketers = np.array(["Tim","John","Jacob","Karl","Gary","Mike"])
runs = np.array([85,59,99,77,67,95])

plt.pie(runs, labels = cricketers, autopct='%1.2f%%')
plt.title("Runs scored in a cricket match")

plt.show()
```

## Line Graph

```python
import matplotlib.pyplot as plt
import numpy as np

months = np.array(["Jan","Feb","Mar","Apr","May","Jun","July","Aug","Sept","Oct","Nov","Dec"])
rain = np.array([60,55,40,45,70,80,85,70,62,79,36,67])

plt.plot(months,rain)

plt.title("Rainfall in a year")
plt.xlabel("Months")
plt.ylabel("Rainfall")

plt.show()
```

## Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

arr = np.array([10,50,34,67,21,7,59,62,45,48,10,8,41,32,66,59,18,79,77,65])

plt.hist(arr, bins = [0,20,40,60,80])

plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

## Scatter Plot

```python
import matplotlib.pyplot as plt

team1_Score = [25,47,34,38,27,40,42,18]
team2_Score = [7,22,40,29,27,10,19,31]

scoreRange = [5,10,15,20,25,30,35,40]

plt.scatter(team1_Score, scoreRange, color = 'r')
plt.scatter(team2_Score, scoreRange, color = 'b')

plt.title("Scatter Plot")
plt.xlabel("Team 1 Score")
plt.ylabel("Team 2 Score")

plt.show()
```
